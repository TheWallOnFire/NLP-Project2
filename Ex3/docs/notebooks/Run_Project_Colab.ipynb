### Step 2: Install and Start Ollama

```python
!curl -fsSL https://ollama.com/install.sh | sh

import os
import subprocess
import threading
import time

env = os.environ.copy()
env["OLLAMA_HOST"] = "0.0.0.0"

def run_ollama_serve():
    subprocess.Popen(["ollama", "serve"], env=env)

threading.Thread(target=run_ollama_serve, daemon=True).start()
time.sleep(5)  # Wait for server to start
```

### Step 5: Expose API via Localtunnel
Google Colab blocks direct port access. We will use `localtunnel` to get a public URL.
**Note:** Localtunnel will ask you for a password (Endpoint IP). Copy the IP printed by the cell below.

```python
!npm install -g localtunnel

import urllib
import subprocess

print("COPY THIS IP ADDRESS (Localtunnel Password):", urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip())

tunnel = subprocess.Popen(["lt", "--port", "8000"], stdout=subprocess.PIPE)
url = tunnel.stdout.readline()
print("\nYOUR PUBLIC API URL:", url.decode('utf-8').strip())
print("Click the link above, enter the IP address, and then add '/docs' to the end of the URL to see the Swagger UI!")
```
